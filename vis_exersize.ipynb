```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/zgulde/018f5d601efc79cb741a7aba92f51d29/raw/c08f1c09f605b43e9a354890f9dd68f296e76dc4/students.csv'
df = pd.read_csv(url)
```

```python
df.head()
```


Create a scatter plot of syntax_errors and ml_coffee. 

Create a scatter plot of p_days_absent and ml_coffee. 


```python
x = df.syntax_errors
y = df.ml_coffee

plt.xlabel("number of syntax errors per time unit")
plt.ylabel("Coffee Consumed in Milileters")
plt.title("Syntax Errors Increase as Coffee Consumption Increases")
plt.scatter(x, y)
plt.show()
```

```python
x = df.p_days_absent
y = df.ml_coffee


plt.xlabel("Percent of Absences pre Time Period")
plt.ylabel("Coffee Consumed in Milileters")
plt.title("Coffee Consumption has no Impact on Absences")
plt.scatter(x, y)
plt.show()
```
