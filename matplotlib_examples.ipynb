# Example for Applying a Formula Across a Range of Numbers

```python
# 1) create list for range using list comprehension 
range_ten = [r for r in range(1,11)]
range_ten
```

```python
# 2) create a function that accepts a number and returns the number with the formula applied to it

x = 4

def formula(x):
    '''
    accepts an number inputand returns that number squared.
    '''
    
    return x ** 2

formula(x)
```

```python
# 3) create a list comprehension that funnels numbers from your range list through your function

numbers_squared = [formula(x) for x in range_ten]

numbers_squared
```

# Example Plotting Coordinates 

```python
# 0) import matplotlib

import matplotlib.pyplot as plt
```

```python
# 1) assign coordinates

x_axis = [r for r in range(1,11)]  # list of numbers from one to ten

y_axis = numbers_squared  # list of squared numbers from above example

print(x_axis)
print(y_axis)
```

```python
# 2) create graph

plt.plot(x_axis,y_axis)  # create graph object
plt.show()  # display graph object
```

# Example Title

```python
plt.plot(x_axis,y_axis) 
 
plt.title("Exponential Growth Curve ") # create title
 
plt.show() # additions to graph must be made before show is called to appear on the graph
```

# Example Axis labels

```python
plt.plot(x_axis,y_axis) 
 
plt.title("Exponential Growth Curve ") 

plt.xlabel('X-Axis') # create lable for x-axis
plt.ylabel('Y-Axis') # create lable for y-axis

plt.show()
```
